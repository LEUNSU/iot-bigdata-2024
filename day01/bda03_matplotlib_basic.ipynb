````python
# Matplotlib 사용 / 필수 차트 사용 시 항상 사용
import matplotlib.pyplot as plt # plt로 줄여씀
from matplotlib import rcParams, font_manager, rc
import warnings

# 맑은고딕으로 변경
font_path = 'C:/Windows/Fonts/malgun.ttf'
font = font_manager.FontProperties(fname=font_path, size=10).get_name()

# warning 표시 끄기
warnings.simplefilter('ignore') # 경고메시지 무시

# 기본폰트 설정
plt.rcParams['font.family'] = font

# 기본폰트 사이즈 설정
plt.rcParams['font.size'] = 12

# 기본차트 사이즈 변경
plt.rcParams['figure.figsize'] = (12, 5)

# 기본그리드 표시
plt.rcParams['axes.grid'] = True

# 한글화 후 마이너스 표시 네모를 정상출력
plt.rcParams['axes.unicode_minus'] = False
````

````python
import numpy as np
import seaborn as sns # seaborn 추가로 코딩

# style => dark, darkgrid, white, whitegrid, ticks
sns.set(font='Malgun Gothic', rc={'axes.unicode_minus':False}, style='')
````

````python
# 넘파이로 더미데이터 생성
# random == rand 양의 실수, randn
data = np.random.random(50)  # 정수로 랜덤 표준정규분포값 50개를 만듦
data
````

````python
# 정수로 만들기
np.random.randint(1, 20, size = 50)
````

````python
## 차트 생성
plt.plot(np.random.randn(50), label='겨울기온')
plt.title('변화값')
plt.legend(loc='upper right')
plt.show()
````

````python
## 시본 차트 생성
sns.set_style('darkgrid')
plt.title('변화값')
plt.legend(loc='upper right')
plt.show()
````

````python
## 스캐터차트
x = np.random.randint(-100, 100, 50) # -100~100 사이에 50개만 랜덤
y = np.random.randint(-200, 200, 50)
size = np.random.randint(-200, 200, 50)
color = np.random.randint(0, 100, 50)

plt.scatter(x, y, s=size, c=color, alpha=0.5)
plt.colorbar()
plt.show()
````

````python
##  파이차트
size = [2441, 2312, 1031, 1233]
label = ['A', 'B', 'AB', 'O']
color = ['darkmagenta', 'deeppink', 'hotpink','pink']

plt.axis('equal') # 가로축, 세로축 스케일 동일
plt.pie(size, labels=label, autopct='%.1f', explode=(0,0,0.1,0), colors=color)
plt.legend(loc='lower left')
plt.title('혈액형별 인구율')
plt.show()
````
